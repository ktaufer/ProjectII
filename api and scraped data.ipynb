```python
import flickrapi
from flickr import api_key, api_secret
import pandas as pd
from pprint import pprint

flickr = flickrapi.FlickrAPI(api_key, api_secret, format = 'parsed-json')
```

```python
sightings_df = pd.read_csv('metadata.csv')
print(len(sightings_df))
sightings_df.columns
```

```python
clean_df = sightings_df[['Recording_ID', 'English_name', 'Species', 'Country','Spectrogram_med', 'Date', 'Url', 
                         'Latitude', 'Longitude', 'Vocalization_type', 'Recordist']]
clean_df.head()
```

```python
# remove data labeled with dubious taxonomy
clean_df = clean_df[clean_df['Species'] != 'Sonus naturalis']
len(clean_df)
```

```python
# remove rows with missing coordinates
clean_df = clean_df.dropna()

len(clean_df)
clean_df.head()
```

```python
# get distinct countries in dataset w/ number of sightings
countries = clean_df.Country.value_counts()
countries.head()
```

```python
#get a list of unique species names in dataset
species = clean_df.Species.unique()
species
```

```python
photos = []

for bird in species:
    photo = flickr.photos.search(text = bird, per_page = '6', content_type = '1', media = 'photos', extras = 'url_m')
    photos.append(photo)
pprint(photos)
```

```python
image_list = []
for item in photos:
    list = item['photos']['photo']
    image_list.append(list)

image1_list = []
image2_list = []
image3_list = []
image4_list = []
image5_list = []
image6_list = []

for item in image_list:
    url0 = item[0]['url_m']
    image1_list.append(url0)
    url1 = item[1]['url_m']
    image2_list.append(url1)
    url2 = item[2]['url_m']
    image3_list.append(url2)
    url3 = item[3]['url_m']
    image4_list.append(url3)
    url4 = item[4]['url_m']
    image5_list.append(url4)
    url5 = item[5]['url_m']
    image6_list.append(url5)
    
len(image1_list)
```

```python
species_df = pd.DataFrame({'Species': species, 'Image1' : image1_list, 'Image2' : image2_list, 'Image3' : image3_list, 
                          'Image4' : image4_list, 'Image5' : image5_list, 'Image6' : image6_list})
species_df.to_csv('images.csv')
species_df.head()
```

```python
clean_df['mp3_url'] = ''
import time

for index, row in clean_df.iterrows():
    url = 'http:'+ row['Url']
    response = requests.get(url)
    time.sleep(2)
    soup = bs(response.content, 'html.parser')
    audio_player = soup.find('div', class_="jp-type-single")
    try:
        link = audio_player.get('data-xc-filepath')
        clean_df.loc[index, 'mp3_url'] = link
        print(f"Processing Record {index} of {len(clean_df)} for record {row['Recording_ID']}.")
    except AttributeError:        
        print('link not found, skipping')
        pass
            
    print('===========================')
    
clean_df.head()
```
